```python
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
dir(digits)
```

```python
# plt.gray()
for i in range(15):
    plt.matshow(digits.images[i])
```

```python
digits.target_names[0:10]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(digits.data, digits.target, test_size=0.15)
```

```python
len(Y_test)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver = "newton-cg", multi_class="auto")
```

```python
model.fit(X_train, Y_train)
```

```python
model.score(X_test,Y_test)
```

```python
plt.imshow(digits.images[99])
```

```python
model.predict([digits.data[99]])
```

```python
y_predicted = model.predict(X_test)
from sklearn.metrics import confusion_matrix
#confusion matrix takes two parameters: First will be the true/confirmed values and second will be the predicted values
cm = confusion_matrix(Y_test,y_predicted)
cm
```

```python
# visualizing the cunfusion matrix
import seaborn as sb
plt.figure(figsize=(10,7))
sb.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
