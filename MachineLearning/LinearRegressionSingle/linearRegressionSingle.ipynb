```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
file = pd.read_csv("homeprices.csv")
file
```

```python
#magic command. Used to run terminal commands in Jupyter Notebook
%matplotlib inline
plt.xlabel('area(sqr ft.)')
plt.ylabel('price(USD)')
plt.scatter(file.area, file.price, color='green', marker='+')
```

```python
rg_model=linear_model.LinearRegression()
rg_model.fit(file[['area']],file[['price']])
```

```python
rg_model.coef_
```

```python
rg_model.intercept_
```

```python
#y=(m*x)+b
#where 'y' is the area which is unknown, 'm' is the coefficient i.e:135.78767123, 'x' is the entered price i.e:3300 
# 'b' is the intercept i.e: 180616.43835616 So plugging in the values gives up?
(135.78767123*3300)+180616.43835616
```

```python
#magic command. Used to run terminal commands in Jupyter Notebook
%matplotlib inline
plt.xlabel('area(sqr ft.)')
plt.ylabel('price(USD)')
plt.scatter(file.area, file.price, color='green', marker='+')
plt.plot(file.area, rg_model.predict(file[['area']]),color='orange')
```

```python
rg_model.predict([[3300]])
```

```python
file2= pd.read_csv('areas.csv')
file2.head(1)
```

```python
#Storing the predicted values in a variable
prices=rg_model.predict(file2)
```

```python
#Creating a new column in file and storing values in it
file2['prices']=prices
file2
```

```python
file2.to_csv('prediction.csv', index = False)
```
